```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
amount = 24
x = np.random.randint(0, 100, [amount])
r = np.random.randint(-2, 2, [amount])
y = x * 0.3 + r
```

```python
plt.scatter(x, y, s=10)
plt.grid()
plt.show()
```

```python
# width = np.random.random()
width = 1
```

```python
y_mean = y.mean()
alpha = 0.001
```

```python
for _ in range(20):
    predict_mean = np.mean(x * width)
    # print(predict_mean)
    error = (predict_mean - y_mean) ** 2
    dir_and_amt = (predict_mean - y_mean)

    # direct = -1 if (predict_mean - y_mean) < 0 else 1
    direct = np.sign(predict_mean - y_mean)
    # width -= dir_and_amt * alpha
    width -= error * direct * alpha
    print(y_mean, predict_mean, width)
```
