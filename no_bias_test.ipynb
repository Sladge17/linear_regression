```python
epochs = 10
alpha = 0.00001
for _ in range(epochs):
    delta = 0
    for i in range(len(x)):
        predict = x[i] * weight
        delta += (predict - y[i]) * x[i]
    weight -= delta * alpha
```

```python
x_pred = np.array((np.min(x), np.max(x)))
y_pred = x_pred * weight
```

```python
plt.scatter(x, y, s=10)
plt.plot(x_pred, y_pred, color="#00FF00")
plt.grid()
plt.show()
```
